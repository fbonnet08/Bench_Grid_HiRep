```python
# ---------- Imports ------------------------------------
import matplotlib
import sys, os, time
import seaborn
import cirq
import torch
import pandas, seaborn
import numpy
import scipy
import random
import matplotlib.pyplot

nx = 30
ny = 30
mu = 0
sigma = 50
```

```python
A=numpy.zeros([nx, ny], dtype='float32')
A=numpy.random.random((nx, ny))
matplotlib.pyplot.matshow(A)
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()
```

```python
panda_array=pandas.DataFrame(A)

print(panda_array)

```

```python
# -------------------------------------------------------
print("Gaussian numbers matrix representation ...")
# -------------------------------------------------------
rnd_numbers = numpy.zeros([nx,ny], dtype='float32')
print ("rnd_numbers: ", rnd_numbers[:,:])
for i in range(nx):
    for j in range(ny):
        rnd_numbers[i][j] = random.gauss(mu=mu,sigma=sigma)

print ("rnd_numbers: ", rnd_numbers[:,:])
matplotlib.pyplot.imshow(rnd_numbers[:,:])
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()
```
